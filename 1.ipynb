```python
import random
def printBoard(board):
    for i in range(3):
        for j in range(3):
            if board[i][j]==1:
                print("X",end="\t")
            elif(board[i][j]==-1):
                print("O",end="\t")
            else:
                print("-",end="\t")
        print("\n")

def end(board):
    for i in range(3):
        if(board[i][0]==board[i][1]==board[i][2]!=0):
            return board[i][0]
    for i in range(3):
        if(board[0][i]==board[1][i]==board[2][i]!=0):
            return board[0][i]
    if(board[0][0]==board[1][1]==board[2][2]!=0):
        return board[0][0]
    if(board[0][2]==board[1][1]==board[2][0]!=0):
        return board[0][2]
    computer_places =0
    user_places =0
    for i in range(3):
        for j in range(3):
            if board[i][j]==1:
                computer_places+=1
            if board[i][j]==-1:
                user_places+=1
    if computer_places+user_places==9:
        return 0
    return None

def minimax(board,chance,cost):
    if(end(board)==None):
        board_copy=board.copy()
        posi_empty=[]
        ansIndex=0
        
        for i in range(3):
            for j in range(3):
                if board_copy[i][j]==0:
                    index=(i*3)+j
                    posi_empty.append(index)
        if chance==0:
            nextChance=1
            result=1
            costResult=1000000
        else:
            result=-1
            costResult=1000000
            nextChance=0
        for i in range(len(posi_empty)):
            index=posi_empty[i]
            indexi=index//3
            indexj=index%3
            if(chance==1):
                board_copy[indexi][indexj]=1
                ans=minimax(board_copy,nextChance,cost+1)
                if(ans[0]>result):
                    result=ans[0]
                    costResult=ans[1]
                    ansIndex=index 
            else:
                board_copy[indexi][indexj]=-1
                ans=minimax(board_copy,nextChance,cost+1)
                if(ans[0]<result):
                    result=ans[0]
                    costResult=ans[1]
                    ansIndex=index 
            board_copy[indexi][indexj]=0
        return (result,costResult,ansIndex)
        
    else:
        return (end(board),cost)

if __name__=='__main__':
    while(True):
        board=[]
        for i in range(3):
            indiarray=[]
            for j in range(3):
                indiarray.append(0)
            board.append(indiarray)
        chance=int(input("Enter who is starting first \n 0 for you, 1 for computer   "))
        print("\n")
        count=0
        while(end(board) is None):
            if(chance==1):
                result=minimax(board,chance,0)
                index=result[2]
                indexi=index//3
                indexj=index%3
                board[indexi][indexj]=1
                count+=1
                print("Computer's Turn")
                printBoard(board)
                print("\n")
            else:
                print("User Turn")
                position=int(input("Enter the index no between 1 and 9   "))
                indexi=(position-1)//3
                indexj=(position-1)%3
                if(board[indexi][indexj]!=0):
                    print("The index is already played")
                    continue
                board[indexi][indexj]=-1
                count+=1
                printBoard(board)
                print("\n\n")
            if(chance==1):
                chance=0
            else:
                chance=1
        ans=end(board)
        if(ans==0):
            print("Draw")
        elif(ans==1):
            print("Computer won")
        else:
            print("You won")
        check=(input("Enter 0 to start new game press any other keystroke to exit   "))
        if(check!='0'):
            break
```
