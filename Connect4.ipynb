```python
import math
def createBoard():
    array=[]
    for i in range(6):
        indiarray=[]
        for j in range(7):
            indiarray.append(0)
        array.append(indiarray)
    return array

def printBoard(board):
    for i in range(6):
        print(*board[i],sep='\t')
def isValidColumn(board,col):
    if board[0][col]==0:
        return True
    return False

def getNextRow(board,col):
    i=5
    while i>=0:
        if board[i][col]==0:
            return i
        i-=1
        
def winningMove(board,move):
    for i in range(6):
        for j in range(4):
            if board[i][j]==move and board[i][j+1]==move and board[i][j+2]==move and board[i][j+3]==move:
                return True
    for i in range(3):
        for j in range(7):
            if board[i][j]==move and board[i+1][j]==move and board[i+2][j]==move and board[i+3][j]==move:
                return True
    for i in range(3,6):
        for j in range(4):
            if board[i][j]==move and board[i-1][j+1]==move and board[i-2][j+2]==move and board[i-3][j+3]==move:
                return True
    for i in range(3):
        for j in range(4):
            if board[i][j]==move and board[i+1][j+1]==move and board[i+2][j+2]==move and board[i+3][j+3]==move:
                return True

def getValidColumns(board):
    validColumns=[]
    for i in range(7):
        if board[0][i]==0:
            validColumns.append(i)
    return validColumns

def getTerminalNode(board):
    return winningMove(board,-1) or winningMove(board,1) or len(getValidColumns(board))==0

def checkWindow(check,player):
    score=0
    if player==1:
        oppPlayer=0
    else:
        oppPlayer=1
    if check.count(player)==4:
        score+=100
    elif check.count(player)==3 and check.count(0)==1:
        score+=5
    elif check.count(player)==2 and check.count(0)==2:
        score+=2
    if check.count(oppPlayer)==3 and check.count(0)==1:
        score-=4
    return score

def getScore(board,player):
    score=0
    centerArray=[]
    for i in range(6):
        centerArray.append(board[i][3])
    centerCount=centerArray.count(player)
    score+=centerCount*3
    for i in range(6):
        for j in range(4):
            check=board[i][j:j+4]
            score+=checkWindow(check,player)
    for j in range(7):
        colArray=[]
        for i in range(6):
            colArray.append(board[i][j])
        for i in range(3):
            check=colArray[i:i+4]
            score+=checkWindow(check,player)
    for i in range(3):
        for j in range(4):
            check=[board[i+k][j+k] for k in range(4)]
            score+=checkWindow(check,player)
    for i in range(3,6):
        for j in range(4):
            check=[board[i-k][j+k] for k in range(4)]
            score+=checkWindow(check,player)
    return score
            
def minimax(board,depth,alpha,beta,maxPlayer):
    validColumns=getValidColumns(board)
    isTerminalNode=getTerminalNode(board)
    if isTerminalNode or depth==0:
        if isTerminalNode:
            if winningMove(board,-1):
                return (None,-1000000)
            elif winningMove(board,1):
                return (None,1000000)
            else:
                return (None,0)
        else:
            return (None,getScore(board,1))
    if maxPlayer:
        value=-math.inf
        column=-1
        for col in validColumns:
            row=getNextRow(board,col)
            board[row][col]=1
            newScore=minimax(board,depth-1,alpha,beta,False)[1]
            board[row][col]=0
            if(newScore>value):
                value=newScore
                column=col
            alpha=max(value,alpha)
            if alpha>=beta:
                break
        return (column,value)
    else:
        value=math.inf
        column=-1
        for col in validColumns:
            row=getNextRow(board,col)
            board[row][col]=-1
            newScore=minimax(board,depth-1,alpha,beta,True)[1]
            board[row][col]=0
            if(newScore<value):
                value=newScore
                column=col
            beta=min(value,beta)
            if alpha>=beta:
                break
        return (column,value)
    

if __name__=='__main__':
    board=createBoard()
    printBoard(board)
    print("\n")
    start=int(input("Enter 0 for you to start and enter 1 for computer to start: "))
    print("\n")
    gameOver=False
    winner=0
    i=0
    while not gameOver and i<42:
        if start==0:
            column=int(input("Enter the column number where you want to place the piece: "))-1
            print("\n")
            if isValidColumn(board,column):
                row=getNextRow(board,column)
                board[row][column]=-1
                printBoard(board)
                print("\n")
                if winningMove(board,-1):
                    gameOver=True
                    winner=-1
            else:
                continue
        if not gameOver:
            col,score=minimax(board,5,-math.inf,math.inf,True)
            if isValidColumn(board,col):
                row=getNextRow(board,col)
                board[row][col]=1
                print("Computer's Turn\n")
                printBoard(board)
                print("\n")
                if winningMove(board,1):
                    gameOver=True
                    winner=1
                start=0
            else:
                start=1
                continue
        i+=2
    if winner==1:
        print("Computer Won the game")
    if winner==-1:
        print("You won the game")
    if winner==0:
        print("Draw")
```
