```python
import random
import math
def printBoard(board):
    for i in range(3):
        for j in range(3):
            if board[i][j]==1:
                print("X",end="\t")
            elif(board[i][j]==-1):
                print("O",end="\t")
            else:
                print("-",end="\t")
        print("\n")

def end(board):
    for i in range(3):
        if(board[i][0]==board[i][1]==board[i][2]!=0):
            #return (board[i][0]==1?100000:-100000)
            return 100000 if board[i][0]==1 else -100000
    for i in range(3):
        if(board[0][i]==board[1][i]==board[2][i]!=0):
            #return board[0][i]==1?100000:-100000
            return 100000 if board[0][i]==1 else -100000
    if(board[0][0]==board[1][1]==board[2][2]!=0):
        #return board[0][0]==1?100000:-100000
        return 100000 if board[0][0]==1 else -100000
    if(board[0][2]==board[1][1]==board[2][0]!=0):
        #return board[0][2]==1?100000:-100000
        return 100000 if board[0][2]==1 else -100000
    computer_places =0
    user_places =0
    for i in range(3):
        for j in range(3):
            if board[i][j]==1:
                computer_places+=1
            if board[i][j]==-1:
                user_places+=1
    if computer_places+user_places==9:
        return 0
    return None

def getBest(board,chance,alpha,beta):
    if(end(board)==None):
        value=0
        ansIndex=0
        board_copy=board.copy()
        posi_empty=[]
        costminmax={}
        for i in range(3):
            for j in range(3):
                if board_copy[i][j]==0:
                    index=(i*3)+j
                    posi_empty.append(index)
        if chance==0:
            nextChance=1
            value=math.inf
        else:
            value=-math.inf
            nextChance=0
        for i in range(len(posi_empty)):
            index=posi_empty[i]
            indexi=index//3
            indexj=index%3
            if(chance==1):
                board_copy[indexi][indexj]=1
                ans=getBest(board_copy,nextChance,alpha,beta)
                board_copy[indexi][indexj]=0
                if(ans[1]>value):
                    value=ans[1]
                    ansIndex=index   
                alpha=max(value,alpha)
                if(alpha>=beta):
                    break
            else:
                board_copy[indexi][indexj]=-1
                ans=getBest(board_copy,nextChance,alpha,beta)
                board_copy[indexi][indexj]=0
                if(ans[1]<value):
                    value=ans[1]
                    ansIndex=index   
                beta=min(value,beta)
                if(alpha>=beta):
                    break
        return (ansIndex,value)    
    else:
        return (None,end(board))

if __name__=='__main__':
    while(True):
        board=[]
        for i in range(3):
            indiarray=[]
            for j in range(3):
                indiarray.append(0)
            board.append(indiarray)
        chance=int(input("Enter who is starting first \n 0 for you, 1 for computer   "))
        print("\n")
        count=0
        while(end(board) is None):
            if(chance==1):
                result=getBest(board,chance,-math.inf,math.inf)
                index=result[0]
                indexi=index//3
                indexj=index%3
                board[indexi][indexj]=1
                count+=1
                print("Computer's Turn")
                printBoard(board)
                print("\n")
            else:
                print("User Turn")
                position=int(input("Enter the index no between 1 and 9   "))
                indexi=(position-1)//3
                indexj=(position-1)%3
                if(board[indexi][indexj]!=0):
                    print("The index is already played")
                    continue
                board[indexi][indexj]=-1
                count+=1
                printBoard(board)
                print("\n\n")
            if(chance==1):
                chance=0
            else:
                chance=1
        ans=end(board)
        if(ans==0):
            print("Draw")
        elif(ans==100000):
            print("Computer won")
        else:
            print("You won")
        check=(input("Enter 0 to start new game press any other keystroke to exit   "))
        if(check!='0'):
            break       
```
