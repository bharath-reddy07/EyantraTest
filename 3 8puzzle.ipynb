```python
from queue import PriorityQueue
def checkEqual(tempState,finalState):
    for i in range(3):
        for j in range(3):
            if(tempState[i][j]!=finalState[i][j]):
                return False
    return True

def getStates():
    State=[]
    for i in range(3):
        temp=[]
        print("Enter values of "+str(i+1)+" row, enter - for empty space")
        for j in range(3):
            number=input()
            temp.append(number)
        State.append(temp)
    return State

def printState(tempState):
    for i in range(3):
        for j in range(3):
            print(tempState[i][j],end='\t')
        print('\n')
def geth(tempState,finalState):
    count=0
    for i in range(3):
        for j in range(3):
            if(finalState[i][j]=='-'):
                continue
            else:
                if(finalState[i][j]!=tempState[i][j]):
                    count+=1
    return count
def getBestState(tempState,g,finalState,visitedStates,queue,fullnodeNumber,parentdict,nodeNumber):
    indexi=0
    indexj=0
    flag=0
    f=100
    for i in range(3):
        for j in range(3):
            if(tempState[i][j]=='-'):
                indexi=i
                indexj=j
                flag=1
                break
            if(flag==1):
                break
    tempi=0
    tempj=0
    parentno=nodeNumber
    for i in range(4):
        if(i==0):
            tempi=indexi+1
            tempj=indexj
        if(i==1):
            tempi=indexi-1
            tempj=indexj
        if(i==2):
            tempi=indexi
            tempj=indexj+1
        if(i==3):
            tempi=indexi
            tempj=indexj-1
        alterState=[row[:] for row in tempState]
        if((tempi>=0 and tempi<=2) and (tempj>=0 and tempj<=2)):
            fullnodeNumber+=1
            tempvalue=alterState[tempi][tempj]
            alterState[tempi][tempj]=alterState[indexi][indexj]
            alterState[indexi][indexj]=tempvalue
            if(alterState not in visitedStates):
                h=geth(alterState,finalState)
                tempf=h+g
                queue.put((tempf,h,g+1,fullnodeNumber,[row[:] for row in alterState]))
                parentdict[fullnodeNumber]=(parentno,[row[:] for row in alterState])
    return fullnodeNumber
if __name__=="__main__":
    print("Enter the starting state of the game")
    initialState=getStates()
    print("Enter the final state of the game")
    finalState=getStates()
    tempState = [row[:] for row in initialState]
    visitedStates=[]
    queue=PriorityQueue()
    queue.put((0,0,0,0,tempState))
    parentdict={}
    ansList=[]
    fullnodeNumber=0
    while not queue.empty():
        f,h,g,nodeNumber,tempState=queue.get()
        visitedStates.append(tempState)
        if(checkEqual(tempState,finalState)):
            while(nodeNumber!=0):
                temp=parentdict[nodeNumber]
                ansList.append(temp[1])
                nodeNumber=temp[0]
            ansList.append(initialState)
            break
        fullnodeNumber=getBestState(tempState,g,finalState,visitedStates,queue,fullnodeNumber,parentdict,nodeNumber)
    i=len(ansList)-1
    j=0
    while(i>=0):
        print("State after iteration :"+str(j))
        print("\n")
        printState(ansList[i])
        i-=1
        j+=1
        print("\n\n")
```
